```python
trans = transforms.Compose([transforms.CenterCrop((220, 220)), transforms.ToTensor()])
lfw = TripletImageDataset(PATH, transform=trans)
loader = DataLoader(lfw, batch_size=3)
```

```python
anchor, positive, negative = next(iter(loader))
```

## Model 
  - Resnet

```python
from torchvision.models import resnet34
from utils import to_var
```

```python
resnet = resnet34(num_classes=128)
```

```python
resnet.cuda()
```

## Loss
 - Triplet loss


```python
embeddings_anchor = resnet(to_var(anchor))
embeddings_positive = resnet(to_var(positive))
embeddings_negative = resnet(to_var(negative))
```

```python
from torch import nn
```

```python
triplet_criterion = nn.TripletMarginLoss(margin=.2)
```

```python
triplet_criterion(embeddings_anchor, embeddings_positive, embeddings_negative)
```

## Training
 - pass batch of anchor, positive and negative samples through model
 - find hard positive and hard negative pairs (maybe margin handles this on it's own?)
